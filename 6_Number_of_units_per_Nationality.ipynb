Find the number of apartments per nationality that are owned by people under 30 years old.


Output the nationality along with the number of apartments.


Sort records by the apartments count in descending order.

```python
import pandas as pd
import numpy as np
```

```python
airbnb_units = pd.read_csv("CSV/airbnb_units.csv")
airbnb_units.head()
```

```python
airbnb_hosts = pd.read_csv("CSV/airbnb_hosts.csv")
columns_to_drop = ['Unnamed: 4', 'Unnamed: 5','Unnamed: 6']
airbnb_hosts = airbnb_hosts.drop(columns=columns_to_drop)
airbnb_hosts.head()
```

```python
merged = pd.merge(airbnb_units, airbnb_hosts, on='host_id')
merged
```

```python
merged = merged[(merged['age'] < 30) &(merged['unit_type']=='Apartment')]
merged
```

```python
result = merged.groupby(['nationality'])['unit_id'].nunique().to_frame(
'apartment_count').reset_index().sort_values('apartment_count')
result
```
