Find the review_text that received the highest number of  'cool' votes.
Output the business name along with the review text with the highest numbef of 'cool' votes.

```python
import pandas as pd
import numpy as np
```

```python
yelp_reviews = pd.read_csv("CSV/yelp_reviews.csv")
yelp_reviews
```

```python
result_cool = yelp_reviews[yelp_reviews['cool'] ==yelp_reviews['cool'].max()][['business_name','review_text','cool', 'funny', 'useful']]
result_cool
```

```python
result_funny = yelp_reviews[yelp_reviews['funny'] ==yelp_reviews['funny'].max()][['business_name','review_text','funny', 'cool', 'useful']]
result_funny
```
