Highest Energy Consumption

Find the date with the highest total energy consumption from the Meta/Facebook data centers. Output the date along with the total energy consumption across all data centers.

```python
import pandas as pd
import numpy as np
```

```python
fb_eu_energy = pd.read_csv("../CSV/fb_eu_energy.csv")
fb_eu_energy = fb_eu_energy.iloc[:, :2]
fb_eu_energy
```

```python
fb_asia_energy = pd.read_csv("../CSV/fb_asia_energy.csv")
fb_asia_energy = fb_asia_energy.iloc[:, :2]
fb_asia_energy
```

```python
fb_na_energy =pd.read_csv("../CSV/fb_na_energy.csv")
fb_na_energy = fb_na_energy.iloc[:, :2]
fb_na_energy
```

```python
df = pd.concat([fb_eu_energy,fb_asia_energy,fb_na_energy])
df
```

```python
consumption = df.groupby('date', as_index=False).sum()
consumption

# Параметр as_index=False указывает, что столбец 'date' 
# не должен быть установлен в качестве индекса в результирующем DataFrame.
```

```python
consumption[consumption.consumption==consumption.consumption.max()][['date','consumption']]
# consumption[consumption['consumption'] == consumption['consumption'].max()][['date','consumption']]
```
