Find all wineries which produce wines by possessing aromas of plum, cherry, rose, or hazelnut

Find all wineries which produce wines by possessing aromas of plum, cherry, rose, or hazelnut. To make it more simple, look only for singular form of the mentioned aromas. HINT: if one of the specified words is just a substring of another word, this should not be a hit, but a miss.


Example Description: Hot, tannic and simple, with cherry jam and currant flavors accompanied by high, tart acidity and chile-pepper alcohol heat.
Therefore the winery Bella Piazza is expected in the results.

```python
import pandas as pd
```

```python
winemag_p1 = pd.read_excel("../CSV/winemag_p1.xlsx")
winemag_p1.head(3)
```

```python
row_1 = winemag_p1.iloc[:1]
row_1
```

```python
dic = {}
for i, j in row_1.items():
    dic[i] = j.values[0]
dic
```

```python
winemag_p1.rename(columns=dic, inplace=True)
winemag_p1.head(2)
```

```python
winemag_p1.drop(0, inplace=True)
winemag_p1.head(3)
```

```python
result = winemag_p1[winemag_p1['description'].
                    str.lower().str.contains('\\bplum\\b|\\bcherry\\b|\\brose\\b|\\bhazelnut\\b')
                    ][['winery']].drop_duplicates()
result
```

Давайте разберем этот код по шагам:

winemag_p1['description']: Этот фрагмент извлекает столбец 'description' из DataFrame winemag_p1. Вероятно, этот столбец содержит описания вин.

.str.lower(): Этот метод применяет функцию lower() ко всем строковым значениям в столбце 'description', чтобы преобразовать текст в нижний регистр. Это делается для того, чтобы обеспечить регистронезависимый поиск.

.str.contains('\\bplum\\b|\\bcherry\\b|\\brose\\b|\\bhazelnut\\b'): Этот метод ищет вхождения текста в столбце 'description', которые содержат слова "plum", "cherry", "rose" или "hazelnut". Обратите внимание, что \\b используется для обозначения границ слова, чтобы избежать частичных совпадений.

winemag_p1[...]: Этот фрагмент фильтрует DataFrame winemag_p1, оставляя только те строки, где условие из предыдущего шага равно True. То есть в результате будут только те строки, в которых описание содержит одно из указанных слов.

[['winery']]: Этот фрагмент извлекает столбец 'winery' из отфильтрованного DataFrame.

.drop_duplicates(): Этот метод удаляет дубликаты из результата, оставляя только уникальные значения в столбце 'winery'.

Таким образом, результатом этого кода будет DataFrame, содержащий уникальные названия винодельний ('winery'), чьи описания включают слова "plum", "cherry", "rose" или "hazelnut".
