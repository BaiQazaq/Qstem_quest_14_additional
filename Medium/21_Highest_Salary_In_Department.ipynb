Highest Salary In Department

Find the employee with the highest salary per department.
Output the department name, employee's first name along with the corresponding salary.

```python
import pandas as pd
import numpy as np
```

```python
employee = pd.read_csv("../CSV/employee.csv")
employee.head()
```

```python
df = employee[employee.groupby('department')['salary'].transform('max') == employee['salary']]
df
```

```python
result = df[['department','first_name','salary']]
result
```

[['department','first_name','salary']] selects the 'department', 'first_name', and 'salary' columns from the filtered rows.
