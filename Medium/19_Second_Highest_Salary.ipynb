Second Highest Salary

Find the second highest salary of employees.

```python
import pandas as pd
import numpy as np
```

```python
employee = pd.read_csv("../CSV/employee.csv")
employee.head()
```

```python
distinct_salary = employee.drop_duplicates(subset = 'salary')
distinct_salary.head()
```

```python
distinct_salary.loc[:, 'rnk'] = distinct_salary['salary'].rank(method='dense', ascending=False)
#distinct_salary['rnk'] = distinct_salary['salary'].rank(method='dense', ascending=False)
distinct_salary.head()
```

```python
result = distinct_salary[distinct_salary.rnk == 2][['salary']]
result
```

Solution Walkthrough
In this walkthrough, we will explore how to find the second highest salary of employees using the Pandas library in Python. We will break down the code step by step to understand the underlying logic and walk through the data manipulation process.

Understanding The Data
Before we begin, we need to make sure we have the necessary data. The code assumes that we have a Pandas DataFrame called "employee" with a column named "salary".

The Problem Statement
We are tasked with finding the second highest salary among all employees in the dataset. To achieve this, we will filter out any duplicate salaries, rank the unique salaries in descending order, and return the salary with a rank equal to 2.

Breaking Down The Code
Let's break down the code step by step:

import pandas as pd
import numpy as np
In the first two lines, we import the necessary libraries: Pandas, which provides data manipulation tools, and NumPy, which provides numerical computing functions.

distinct_salary = employee.drop_duplicates(subset="salary")
Here, we create a new DataFrame called "distinct_salary" by dropping any duplicate rows based on the "salary" column in the "employee" DataFrame. This ensures that we only have unique salaries to work with.

distinct_salary["rnk"] = distinct_salary["salary"].rank(
    method="dense", ascending=False
)
Next, we add a new column called "rnk" to the "distinct_salary" DataFrame. We assign ranks to each salary using the rank function from Pandas. The method='dense' parameter ensures that the ranks are assigned without gaps. The ascending=False parameter means that the salaries will be rank ordered in descending order.

result = distinct_salary[distinct_salary.rnk == 2][["salary"]]
Finally, we create a new DataFrame called "result" by filtering the "distinct_salary" DataFrame for rows where the rank ("rnk") is equal to 2. We select only the "salary" column for the final result.

Bringing It All Together
The complete code is as follows:

import pandas as pd
import numpy as np

distinct_salary = employee.drop_duplicates(subset="salary")
distinct_salary["rnk"] = distinct_salary["salary"].rank(
    method="dense", ascending=False

result = distinct_salary[distinct_salary.rnk == 2][["salary"]]
