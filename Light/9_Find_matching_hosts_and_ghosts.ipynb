Find matching hosts and guests in a way that they are both of the same gender and nationality

```python
import pandas as pd
import numpy as np
```

```python
airbnb_hosts = pd.read_csv("CSV/airbnb_hosts_9.csv")
#airbnb_hosts.drop(airbnb_hosts.columns[[4, 5, 6]], axis=1, inplace=True)
#columns_to_keep = ['host_id', 'nationality', 'gender', 'age']
#airbnb_hosts = airbnb_hosts[columns_to_keep]
airbnb_hosts = airbnb_hosts.iloc[:, 0:4]
airbnb_hosts.head()
```

```python
airbnb_guests = pd.read_csv('CSV/airbnb_guests.csv')
airbnb_guests = airbnb_guests.iloc[:, 0:4]
airbnb_guests.head()
```

```python
merged_df = pd.merge(airbnb_hosts, airbnb_guests,  
                     how='left', left_on=['nationality','gender'], right_on = ['nationality','gender'])
merged_df
```

```python
result = merged_df[['host_id','guest_id']].drop_duplicates()
result
```
